```
import numpy as np
import math
import time
import pandas as pd
```

```
cd=0
ld=0
def get_distance(p1,p2):
    x=p2[0]-p1[0]
    y=p2[1]-p1[1]
    cd=math.sqrt(x*x+y*y)
    return cd
```

```
def get_allDistances(t):
  n = len(t)
  distances = [[0]*n]*n
  t1=time.time()
  for i in range(n):
    for j in range(n):
      distances[i][j]=get_distance(t[i],t[j])
  t2=time.time()
  print('total time taken {}sec:'.format(t2-t1))
  return distances
```
