```
import numpy as np
import math
```

```
cd=0
ld=0
def get_distance(p1,p2):
    x=p2[0]-p1[0]
    y=p2[1]-p1[1]
    cd=math.sqrt(x*x+y*y)
    return cd
```

```
get_distance((1,1),(5,4))
```
